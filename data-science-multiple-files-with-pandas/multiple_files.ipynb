```python
import pandas as pd
import numpy as np
import matplotlib
```

## Merge Alıştırması

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### İç Birleştirme

### İç Birleştirme Alıştırması

```python
inner_merged_df = a_df.merge(b_df, on='Country')
inner_merged_df
```

### Kodunu test et

```python
from nbresult import ChallengeResult

# Testin beklediği değerleri DataFrame'imizden çekerek sonucu oluşturuyoruz
result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=inner_merged_df.isnull().sum().sum()
)

# Sonuçları diske yaz ve testi çalıştırıp çıktısını ekrana bas
result.write()
print(result.check())
```

#### Sol Birleştirme

### Sol Birleştirme Alıştırması

```python
left_merged_df = a_df.merge(b_df, on='Country', how='left')
left_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```
